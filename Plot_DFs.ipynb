```python
from dd_df import triple_plot_df
from dd_df import get_shot_title
import pandas as pd
import matplotlib.pyplot as plt
from dd_df import overview_plot
%load_ext autoreload
%autoreload 2
#
plt.style.use('helvet2')
```

```python
shotdata = [[35156,[[0.5,1.8],[1.8,2.0],[2.0,3.0],[3.0,5.0],[5.0,7.0]], get_shot_title(35156)],
            [35157,[[0.5,1.8],[1.8,2.0],[2.0,3.0],[3.0,5.0],[5.0,7.0]], get_shot_title(35156)],
            [35158,[[0.5,1.7],[1.7,2.0],[2.0,3.0],[3.0,4.0],[4.0,5.0],[5.0,7.0]], '\#35158 Power exhaust with mixed impurities'],
            [35165,[[0.5,1.7],[1.7,3.0],[3.0,4.0],[4.0,4.2],[4.2,5.0],[5.0,6.0]], get_shot_title(35165)]]
```

```python
for entry in shotdata:
    df = pd.read_csv('./Support_Files/df_'+str(entry[0])+'.csv')
    fig, ax = triple_plot_df(df, intervals=entry[1],
                             nev_exponent=20, dotsize=10)
    ax[1].legend(frameon=True, borderpad=0.1, handletextpad=0.1)
    titlestr = r'\#%d'%(entry[0])
    #ax[0].set_title(titlestr, loc='left')
    ax[0].text(0.03, 0.92, titlestr, transform=ax[0].transAxes)
    plt.tight_layout()
    df_image_name = './Images/df_%d.png'%(entry[0])
    plt.savefig(df_image_name, dpi=300)
    plt.show()
    overview_plot(entry[0], entry[1], print_fig=True, titlestr=entry[2])
```
