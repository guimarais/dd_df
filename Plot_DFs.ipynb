```python
from dd_df import triple_plot_df
import pandas as pd
import matplotlib.pyplot as plt
#
plt.style.use('helvet2')
```

```python
df_35156 = pd.read_csv('./Support_Files/df_35156.csv')
interv = [[0,1.8],[1.8,2.0],[2.0,3.0],[3.0,5.0],[5.0,7.0]]
fig, ax = triple_plot_df(df_35156, intervals=interv,
                         nev_exponent=20, dotsize=10)
ax[1].legend(frameon=True, borderpad=0.1, handletextpad=0.1)
ax[0].set_title('\#35156', loc='left')
plt.tight_layout()
plt.savefig('./Images/df35156.png', dpi=300)
plt.show()
```
